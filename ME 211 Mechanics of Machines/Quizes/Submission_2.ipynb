```python
angle = np.linspace( -np.pi ,  4*np.pi , 201 ) 
 
radius = 4
 
xx = radius * np.cos( angle ) 
yy = radius * np.sin( angle )
zz = np.linspace(0,0,100)



fig1 = go.Figure(
            data=[go.Scatter3d(x=[xx[0]], y=[yy[0]],z=[zz[0]],
                            mode="markers",
                            marker=dict(color="red", size=10)),
                go.Scatter3d(x=xx, y=yy,z=zz, name='Path',
                            mode="lines",
                            line=dict(color="blue", width=2))],
            layout=go.Layout(
                scene = dict(
                            xaxis=dict(range=[-5,5], autorange=False),
                            yaxis=dict(range=[-5,5], autorange=False),
                            zaxis=dict(range=[-5,5], autorange=False),
                            aspectratio=dict(x=1, y=1, z=1),
                ),
                title_text='motion respect to e', hovermode="closest",
                updatemenus=[dict(type="buttons",
                                buttons=[dict(label="Play",
                                                method="animate",
                                                args=[None])])]),
            frames=[go.Frame(
                data=[go.Scatter3d(
                    x=[xx[p]],
                    y=[yy[p]],
                    z=[zz[p]],
                    name='Particle',
                    mode="markers",
                    marker=dict(color="red", size=10))])

                for p in range(100)]
)

fig1.show()
```
