```python
import sympy
from sympy.vector.coordsysrect import CoordSys3D
C = CoordSys3D(1)
i, j, k = C.base_vectors()
A=0.3*i+1.4*j+0.1*k
B=0.8*i+(-1.5)*j+(-0.7)*k

cross=A.cross(B)
Area=sympy.sqrt(cross.dot(cross))
print(Area)
```
