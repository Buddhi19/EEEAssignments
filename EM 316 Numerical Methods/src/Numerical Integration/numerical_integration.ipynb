```python
import numpy as np
import matplotlib.pyplot as plt
```

Consider,
\begin{align}
\int_a^be^{-x^2}dx
\end{align}

```python
a = 0
b = 10
n = 10**3

def plot_graph(a,b,n):
    dx = (b-a)/n
    x = np.linspace(a,b,int(b/dx))
    y = np.exp(-x**2)
    plt.figure()
    plt.grid(True)
    plt.plot(x,y)
    plt.show()

plot_graph(a,b,n)
```

```python
plot_graph(0,1,100)
```

```python
def calculate(a,b,n):
    arr = []
    dx = (b-a)/n
    for i in range(n):
        arr.append(np.exp(-(dx*i)**2))

    area = 0
    for i in range(n-1):
        area += (arr[i]+arr[i+1])*dx
    return area
```

```python
print(calculate(0,1,1000))
```
