```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# carrier signal properties
fc = 100
A_c = 1
t = np.linspace(0,np.pi/2,1000)

# carrier signal
c_t = A_c*np.cos(2*np.pi*fc*t)

# message signal properties
fm = 1
A_m = 1
phi = np.pi/4

# message signal
m_t = A_m*np.cos(2*np.pi*fm*t + phi)

# Amplitude Modulated Signal properties
mu = 0.7
AM_t = (1 + mu*m_t)*c_t

# plot the all signals
plt.figure(figsize=(20,8))
plt.subplot(311)
plt.plot(t,c_t)
plt.title('Carrier Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(312)
plt.plot(t,m_t)
plt.title('Message Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(313)
plt.plot(t,AM_t)
plt.title('Amplitude Modulated Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```
